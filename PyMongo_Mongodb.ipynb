```python
import pymongo as pm
import datetime
```

```python
fh = open('C:/Users/dksin/Desktop/Vocabulary_set.csv', "r") # reading the file
```

```python
# creating list
wd_list = fh.readlines()
```

```python
print(wd_list)
```

```python
# removing headings
wd_list.pop(0)
```

```python
vocab_list = [] # empty list

for rawstring in wd_list:
  word, definition = rawstring.split(',', 1) # removing commas
  definition = definition.rstrip()
  vocab_list.append({'word': word, 'definition': definition})

print(vocab_list)
#print(definition)
#print(word)
```

```python
# setting the client

client = pm.MongoClient("mongodb://localhost:27017/")
db = client["vocab"]

# inserting the documents in mongodb id
dbs = client.list_database_names()

vocab_col = db["vocab_list"]

#vocab_col.drop()  # for dropping the database collection


# creating json entry
vocab_dict = {"word":'cryptic', 'definition': 'secret with hidden meaning'}
res = vocab_col.insert_one(vocab_dict)# inserting col
print("inserted_id: ", res.inserted_id)

if "vocab" in dbs:
    print("Database exits")
    
res = vocab_col.insert_many(vocab_list)
#print(res.inserted_ids)
```

# Retriving data : using find_one

```python
data = vocab_col.find_one()
print(data)
```

```python
for data in vocab_col.find({}):
    print(data)
```

```python
# raising exclusion for excluding ids and definition
for data in vocab_col.find({}, {"_id":0, "definition": 0}):
    print(data)
```

```python
data = vocab_col.find_one({'word':'boisterous'})
print(data)
```

# Updating the collection in databases

```python
upd = vocab_col.update_one({'word':'boisterous'},
                           {"$set": {"definition": "rowdy; noisy" }})

print("modified count:",upd.modified_count)
```

```python
upd = vocab_col.update_many({},
    {"$set": {"last_updated UTC":
    datetime.datetime.utcnow().strftime('%Y-%m-%d%H%M$SZ')}})

print("modified count:",upd.modified_count)
```
